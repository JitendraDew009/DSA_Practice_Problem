```python
# Implementation of stack using array data structure
# append -> pushing the data inside the stack
stack = []
stack.append(0)
stack.append(1)
stack.append(2)
stack.append(3)
stack.append(4)
# |------------------------|
# |          4             |
# |          3             |
# |          2             |
# |          1             |
# |          0             |
# |------------------------|

stack
```

```python
# pop the data from the stack
stack.pop()
```

```python
# Implementation of stack using circular deque
from collections import deque
stack = deque()

stack.append(0)
stack.append(1)
stack.append(2)
stack.append(3)
stack.append(4)
```
