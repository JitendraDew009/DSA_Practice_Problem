```python
# This algorithm uses a stack because of the Last-In-First-Out (LIFO) nature of parentheses matching.
# Time Complexity : O(n) (we scan each character once).
# Space Complexity : O(n) (in worst case, all characters are opening brackets).



# Method defination
def isValidParen(string):
    open_param = set(["(", "[", "{"])
    dictionary = {"(" : ")", "[" : "]", "{": "}"}
    stack = []
    for char in string:
        if char in open_param:
            stack.append(char)
        elif stack and char == dictionary[stack[-1]]: # last element of dictionary
            # pop the element from the stack
            stack.pop()
        else:
            return False
    return stack == []                                   

# driver code
string1 = "{[]}"
string2 = "{[[]]}"
if isValidParen(string1):
    print("String contains valid parenthesis")
else:
    print("String contains an invalid parenthesis")
```
