```python
# class of Queue with it's method
class Queue:
    def __init__(self):
        self.buffer = deque()
    
    def enqueue(self, data):
        self.buffer.appendleft(data)

    def isEmpty(self):
        return len(self.buffer) == 0
    
    def size(self):
        return len(self.buffer)
    
    def dequeue(self):
        return self.buffer.pop()
```

```python
q = Queue()
q.enqueue(12)
q.enqueue(6)
q.enqueue(7)
q.enqueue(13) 


```
