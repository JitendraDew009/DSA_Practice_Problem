```python
## Search for an element"15" and if it's present in an array, return the index of that element.
## Suppose the searching element is not present in anarray return -1.

def linearSearch(arr,x):
    for i in range(0,len(arr)):
        if arr[i] == x:
            return i
    return -1

##Driver_code
arr = [2,1,8,9,12,15,11,19]
x = 15
#Function calling
result = linearSearch(arr, x)
print("Searching element is present at the in index", result)
```

```python
## Search for an element"15" and if it's present in an array, return the index of that element.
## Suppose the searching element is not present in anarray return -1.

def linearSearch(arr,x):
    for i in range(0,len(arr)):
        if arr[i] == x:
            return i
    return -1

##Driver_code
arr = [2,1,8,9,12,15,11,19]
x = 16
#Function calling
result = linearSearch(arr, x)
print("Searching element is present at the in index", result)
```
