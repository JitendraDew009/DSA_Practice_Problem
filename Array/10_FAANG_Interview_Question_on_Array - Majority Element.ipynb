```python

from collections import Counter

def majorityElement(nums):
    counts = Counter(nums)
    n = len(nums)
    for key, freq in counts.items():
        if freq > n//2:
            return key
    return key  # no majority element if freq < n//2

nums = [2,2,1,1,1,2,2]
print("Majority Element:", majorityElement(nums))
```

```python
# Approach 2: Hash Table (key : value) Dictionary Based Data Structure
# TC : O(n) - n number of elements
# SC : O(n) - as we know we are storing some key value in pair and it is depend on n number of element in nums. 
from collections import Counter

## Method Defination

def majorityElement(nums):
    counts = Counter(nums)
    print(counts)
    return max(counts.keys(), key = counts.get)

## Driver Code
nums = [2,2,1,1,1,2,2]
result = majorityElement(nums)
print("Majority Element in an array is:", result)
```

```python
from collections import Counter

def mostFrequentElement(nums):
    counts = Counter(nums)
    return max(counts, key=counts.get)

nums = [2,2,1,1,1,1,1,2,2,3,4]
print("Most Frequent Element:", mostFrequentElement(nums))
```

```python
# Approach 3: Boyer Voting Algorithm
# TC : O(n) 
# SC : O(1) 


## Method defination of findCandidate
def findCandidate(nums):
    count = 0 
    candidate = None
    for num in nums:
        if count == 0:
            candidate = num
        count += (1 if num == candidate else -1)
    return candidate

## Method defination of isMajority
def isMajority(nums, candidate):
    cnt = 0
    n = len(nums)
    for i in range(n):
        if nums[i] == candidate:
            cnt += 1
    if cnt > n/2:
        return 1
    else:
        return 0

## Method Defination of printmajorityElement

def printmajorityElement(nums):
    cand = findCandidate(nums)
    if isMajority(nums, cand): # same fucntion but different variable (return 1 means True)
        print("Majority element in an array is:", cand)
    else: # if not true (return 0)
        print("No Majority element exists in an array")
 
## Driver Code
nums = [2,2,1,1,1,2,2]
printmajorityElement(nums)
```
