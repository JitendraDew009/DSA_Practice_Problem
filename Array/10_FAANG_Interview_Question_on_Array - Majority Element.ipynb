```python
# Approach 1: Sorting Method
# TC : O(nlogn) - Sorting a list in Python uses Timsort, which runs in O(n log n) in the average and worst case.

# SC : O(1) - {nums.sort} it doesn’t create a new list.
def sortedmajorityElement(nums):
    nums.sort()  # sort in place
    return nums[len(nums)//2]  # majority max frequency elements appear in the middle 

# driver code
nums = [2,2,1,1,1,2,2,3,3,3,3,3]
result = sortedmajorityElement(nums)
print("Majority Element in an array is:", result)
```

```python
# Approach 2: Hash Table (key : value) Dictionary Based Data Structure
# TC : O(n) - n number of elements
# SC : O(n) - as we know we are storing some key value in pair and it is depend on n number of element in nums. 
from collections import Counter

## Method Defination

def majorityElement(nums):
    counts = Counter(nums)
    print(counts)
    return max(counts.keys(), key = counts.get)

## Driver Code
nums = [2,2,1,1,1,2,2]
result = majorityElement(nums)
print("Majority Element in an array is:", result)
```

```python
# Exploring Counts
counts = {"a": 3, "b": 5, "c": 2}
print(counts.keys())   # dict_keys(['a', 'b', 'c'])
```

```python
print(counts.get("a"))        # 3
print(counts.get("z"))        # None
print(counts.get("z", 0))     # 0 (default value provided)
```
