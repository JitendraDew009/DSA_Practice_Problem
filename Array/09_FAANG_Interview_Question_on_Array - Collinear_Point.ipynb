```python
# Approach 1: Using slope concept
# Time complexity : O(1)
# Space complexity : O(1)

# Method Defination
def iscollinearPoint(x1, x2, x3, y1, y2, y3):
    if ((y2-y1)*(x3-x2) == (y3-y2)*(x2-x1)):
        print("Given points are collinear")

    else:
        print("Given points are non-collinear") 

## Driver Code
x1, x2, x3, y1, y2, y3 = 1, 1, 1, 6, 0, 9
iscollinearPoint(x1, x2, x3, y1, y2, y3)
```

```python
# Approach 2: Using Area of Triangle concept
# Time complexity : O(1)
# Space complexity : O(1)

def iscollinearPoint(x1, x2, x3, y1, y2, y3):
    area = 0.5 * (x1 * (y2-y3) + x2 * (y3-y1) + x3 * (y1-y2))
    if area == 0 :

        print("Given points are collinear")

    else:
        print("Given points are non-collinear") 

## Driver Code
x1, x2, x3, y1, y2, y3 = 1, 2, 3, 6, 0, 9
iscollinearPoint(x1, x2, x3, y1, y2, y3)
```
