```python
from collections import Counter
import heapq

# function_defination
def topKfrequentElement(arr, k):
    if k == len(arr):
        return set(arr)
    
    count = Counter(arr)
    return heapq.nlargest(k, count.keys(), key = count.get)


#Driver_Code
arr = [1,1,1,1,2,2,2,3]
k = 2
#Function_calling
result = topKfrequentElement(arr, k)
print("The top K Frequent element are", result)
```
