```python
## Method Defination sortColors

def sortColors(nums):
    n = len(nums)
    curr = p0  = 0
    p2 = n -1
    while curr <= p2:
        if nums[curr] == 0:
            nums[curr], nums[p0] = nums[p0], nums[curr]
            p0 += 1
            curr += 1        
        
        elif nums[curr] == 2:
            nums[curr], nums[p2] = nums[p2], nums[curr]
            p2 -= 1
        else:
            curr += 1
    return nums


## Driver_code
nums = [2, 0, 2 , 1, 1, 0]
result = sortColors(nums)
print("Sort Colors:",result)
```
