```python
# Overall Time Complexity : O(n)
# Space Complexity : O(1) [Fixed Storage]

# Best time to Buy and Sell Stocks - Synchro
def findmaxProfit(prices):
    min_price = float('inf')
    max_profit = 0

    for i in range(len(prices)):
        if prices[i] < min_price:
            min_price = prices[i]
        
        elif prices[i] - min_price > max_profit:
            max_profit = prices[i] - min_price
    return max_profit

#Driver_Ccode
prices = [7,5,3,6,4,15,20]
## Function  calling
maxProfit = findmaxProfit(prices)
print("The Maximum profit of buy and sell the stocks is:", maxProfit)
```
