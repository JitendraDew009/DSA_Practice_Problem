```python
# Over all Time complexity : O(nlogn)+O(klogn) = O(nlogn)
# Overall Space Compelxity : O(n) space - You store all n points in the heap

from heapq import heappush, heappop
import math
# Function Defination
def get_dist(x,y):
    return math.sqrt(x**2 + y**2) # get_dist is to give the ecludian distance

def kClosest(points, k): # Time complexity : O(nlogn)
    min_heap = []
    n = len(points)
    for i in range(n):
        x = points[i][0]
        y = points[i][1]

        # to insert the elements inside the minheap
        heappush(min_heap, (get_dist(x,y), points[i]))

    result = []

    for i in range(k): # Time complexity : O(klogn)
        # to delete the elements from minheap
        result.append(heappop(min_heap)[1])
    return result



##Driver_code
points = [[3,3],[5,-1],[-2,4]]
k = 2
result = kClosest(points, k)
print("Top K closest point are", result)
```

```python
#  Optimized Max-Heap (using negative distance) Approach
# This version keeps only k elements in the heap at any time, making it faster when k<< n.
# Time Complexity: O(nlogk)
# Space Complexity: O(k)

import math
from heapq import heappush, heappop

def get_dist(x, y):
    return math.sqrt(x**2 + y**2)

def maxkClosest(points, k):
    max_heap = []
    for x, y in points:
        dist = get_dist(x, y)
        heappush(max_heap, (-dist, [x, y]))  # negative distance for max-heap behavior
        if len(max_heap) > k:
            heappop(max_heap)  # remove farthest point

    return [point for (_, point) in max_heap]

# Driver code
points = [[3,3],[5,-1],[-2,4]]
k = 2
print("Top K closest points are", maxkClosest(points, k))
# Faster when k ≪ n
```
