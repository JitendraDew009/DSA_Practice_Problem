```python
#Time complexity : O(n^2)
def selectionSort(arr):
    n = len(arr)
    for i in range(n):
        ## to store the index of the minimum element
        min_idx = i # 58
        for j in range(i+1, n):
            if arr[min_idx] > arr[j] : # 58 > 38 , 38 > 29 , 29 > 11 
                min_idx = j #38 ; 29; 11; 
        # Swap of the element of i and min_idx
        arr[i], arr[min_idx] = arr[min_idx], arr[i] # 58 , 11 = 11 , 58 swap the position of number
    return arr        


# Driver Code
arr = [58, 38, 75, 29, 11, 17, 20, 37]
result = selectionSort(arr)
print("Selection Sort:",result)
```
