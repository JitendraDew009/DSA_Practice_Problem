```python
def selectionSort(arr):
    n = len(arr)
    for i in range(n):
        min_idx = i
        for j in range(i+1, n):
            if arr[j] < arr[min_idx]:
                min_idx = j
        arr[i], arr[min_idx] = arr[min_idx], arr[j]
    return arr        




# Driver Code
arr = [58, 38, 75, 29, 11, 17, 20, 37]
result = selectionSort(arr)
print("Selection Sort:",result)
```
