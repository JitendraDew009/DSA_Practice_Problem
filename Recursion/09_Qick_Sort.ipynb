```python
# Implementation of Qick Sort Algorithm
# Recurrence Relation -> T(n) = T(m-p) + T(q-m) + n
# Time Complexity  
# Best case Scenerio : T(n) = T(n/2) + T(n/2) + n => O(n logn) 
# Worst Case Scenerio:T(n) = T(n-1) + n => O(n^2) 

# Function defination of Partition
def Partition(arr, p, q):
    i = p
    j = i + 1

    # starting element as a pivot element
    pivot = arr[p]
        
    for j in range(i + 1, q + 1):
        if arr[j] <= pivot:
            i += 1
            arr[i], arr[j] = arr[j], arr[i]
    
    #swap between the arr[i] and the pivot element
    arr[i], arr[p] = arr[p], arr[i]
    return i
    
# Function defination of quickSort
def quickSort(arr, p, q):
    if p < q:
        # fucntion calling for partition algorithm
        mid = Partition(arr, p, q)

        #Left side recursive call
        quickSort(arr, p, mid - 1)
    
        #Right side recursive call
        quickSort(arr, mid + 1, q)
    return arr


#Driver Code
arr = [70, 50, 27, 99, 85 , 47, 123, 34, 147]
p = 0
q = len(arr) - 1
result = quickSort(arr, p, q)
print("After applying Quick Sort array, the Sorted array is", result)
```

```python
# Implementation of Random Qick Sort Algorithm
# Recurrence Relation -> T(n) = T(m-p) + T(q-m) + n
# Time Complexity  
# Best case Scenerio : T(n) = T(n/2) + T(n/2) + n => O(n logn) 
# Worst Case Scenerio:T(n) = T(n-1) + n => O(n^2) 

import random 
# Function defination of randomizedPartition
# pivot element is using radom function
def randomizedPartition(arr, p, q):
    random_pivot = random.randrange(p,q)
    # swaping between random_pivot and arr[p] element
    arr[p], arr[random_pivot] = arr[random_pivot], arr[p]
    return Partition(arr, p, q)

# Function defination of Partition
def Partition(arr, p, q):
    i = p
    j = i + 1

    # starting element as a pivot element
    pivot = arr[p]
        
    for j in range(i + 1, q + 1):
        if arr[j] <= pivot:
            i += 1
            arr[i], arr[j] = arr[j], arr[i]
    
    #swap between the arr[i] and the pivot element
    arr[i], arr[p] = arr[p], arr[i]
    return i
    
# Function defination of quickSort
def quickSort(arr, p, q):
    if p < q:
        # fucntion calling for partition algorithm
        mid = randomizedPartition(arr, p, q)

        #Left side recursive call
        quickSort(arr, p, mid - 1)
    
        #Right side recursive call
        quickSort(arr, mid + 1, q)
    return arr


#Driver Code
arr = [70, 50, 27, 99, 85 , 47, 123, 34, 147]
p = 0
q = len(arr) - 1
result = quickSort(arr, p, q)
print("After applying Quick Sort array, the Sorted array is", result)
```
