```python
# Implementation of Qick Sort 
# Function defination of Partition
def Partition(arr, p, q):
    i = p
    j = i + 1

    # starting element as a pivot element
    pivot = arr[p]
    n = len(arr)
    
    for j in range(i + 1, q + 1):
        if arr[j] <= pivot:
            i += 1
            arr[i], arr[j] = arr[j], arr[i]
    
    arr[i], arr[p] = arr[p], arr[i]
    return i
    
# Function defination of quickSort
def quickSort(arr, p, q):
    if p < q:
        # fucntion calling for partition algorithm
        mid = Partition(arr, p, q)

        #Left side recursive call
        quickSort(arr, p, mid - 1)
    
        #Right side recursive call
        quickSort(arr, mid + 1, q)
    return arr


#Driver Code
arr = [70, 50, 27, 99, 85 , 47, 123, 34, 147]
p = 0
q = len(arr) - 1
result = quickSort(arr, p, q)
print("After applying Quick Sort array, the Sorted array is", result)
```
