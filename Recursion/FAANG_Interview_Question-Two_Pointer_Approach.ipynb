```python
# Implementation of Two pointer approach
# Time Complexity: O(n)
# Function defination
def findPair(arr, target):
    l = 0
    r = len(arr) - 1

    for i in range(len(arr) - 1):
        if arr[l] + arr[r] == target:
            return l,r
        elif arr[l] + arr[r] > target: # Decrement r
            r = r - 1
        else: # Increment l
            l = l + 1
    return -1,-1 # No such pair exist that adds up and given target value.

# Driver Code
arr = [20, 40, 50, 75, 120, 145, 200]
target = 90
result = findPair(arr, target)
print(result)
```
