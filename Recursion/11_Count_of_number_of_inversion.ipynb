```python
# Brute force: checks every possible pair -> quadratic growth.
# Recuurence relation -> n(n-1)/2
# Time Complexity -> O(n^2)
# Space Complexity -> O(1)

arr = [70, 50, 60, 10, 20, 30, 80, 15]
inv_count = 0

for i in range(len(arr)):

    for j in range(i+1, len(arr)):
        
        if arr[i] > arr[j]:
            inv_count += 1
        else:
            inv_count += 0
            
print("Number of Inversions in given array:", inv_count)
```

```python
# Merge Sort Based
# Recuurence relation -> 2T(n/2) + n
# Time Complexity -> O(nlogn)
# Space Complexity -> O(n)
# Optimized approach : MergeSort


def merge_and_count(arr, temp, left, mid, right):
    i = left
    j = mid + 1
    k = left
    inv_count = 0

    while i <= mid and j <= right:
        if arr[i] <= arr[j]:
            temp[k] = arr[i]
            i += 1
        else:
            temp[k] = arr[j]
            inv_count += (mid - i + 1)  # count inversions
            j += 1
        k += 1

    while i <= mid:
        temp[k] = arr[i]
        i += 1
        k += 1

    while j <= right:
        temp[k] = arr[j]
        j += 1
        k += 1

    for i in range(left, right + 1):
        arr[i] = temp[i]

    return inv_count


def count_inversion(arr, temp, left, right):
    inv_count = 0
    if left < right:
        mid = (left + right) // 2
        inv_count += count_inversion(arr, temp, left, mid)
        inv_count += count_inversion(arr, temp, mid + 1, right)
        inv_count += merge_and_count(arr, temp, left, mid, right)
    return inv_count


# Driver Code
arr = [70, 50, 60, 10, 20, 30, 80, 15]
n = len(arr)
temp = [0] * n
result = count_inversion(arr, temp, 0, n - 1)
print("Number of Inversions in given array:", result)
```
