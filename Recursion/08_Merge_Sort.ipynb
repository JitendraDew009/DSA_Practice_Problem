```python
# Implementation of Merge Sort
#  function defination of merge Procedure
# n1 + n2 = N
def mergeProcedure(arr, i , mid, j):
    # No. of elements in the left subarray (i, mid)
    n1 = mid - i + 1 
    # No. of elements in the Right subarray (mid+1, j)
    n2 = j - mid
    
    # Initalization of left subarray and Right subarray
    left_Subarray = [0] * n1
    right_Subarray = [0] * n2

    # Copy the elements from an array to the subarrays
    for m in range(n1):
        left_Subarray[m] = arr[i + m]
    for n in range(n2):
        right_Subarray[n] = arr[mid + 1 + n]    
 
    p = 0
    q = 0
    k = i
    # Returning a sorted array
    while p< n1 and q < n2:
        if left_Subarray[p] <= right_Subarray[q]:
            arr[k] = left_Subarray[p] 
            p += 1
        else:
            arr[k] = right_Subarray[q]
            q += 1
        k += 1

        # copy the entire elements from the left subarray
        while p < n1:
            arr[k] = left_Subarray[p]
            p += 1
            k += 1
        # copy the entire elements from the right subarray
        while q < n2:
            arr[k] = right_Subarray[q]
            q += 1
            k += 1

# Function defination 
# Divide and Conquer Approach
# Recurrence Relation ; 2T(N/2) + N = 
# Time Complexity: O(N * logN)
def mergeSort(arr, i, j):
    
    if i < j:
        # Divde
        mid = i + (j-i) // 2
        
        # Conquer
  
        mergeSort(arr, i, mid) # Recursive call for Left subtree
        mergeSort(arr, mid+1, j) # Recursive call for Right subtree

        #Combine
        mergeProcedure(arr,i,mid,j)
    return arr

# Driver
arr = [50, 70, 65, 13, 80, 62, 98, 27]
i = 0
j = len(arr) - 1
result = mergeSort(arr, i, j)
print("Sorted array after applying the merge sort is",result)
```
