```python
# using recursive approach
# function defination
def findfactorial(n):
    #base case condition
    #small problem
    if n == 0 or n == 1:
        return 1
    else:
        # big problem
        # recursion is happening
        return n * findfactorial(n-1) 

## driver_code
n = 8
result = findfactorial(n)
print(f"The factorial of {n} is {result}")
```

```python
# using iterative approach
# function defination
def findfactorial(n):
    result = 1
    if n == 0 or n == 1:
        return 1
    else:
        # iterative approach 
        for i in range(2, n+1):
            result = result * i
        return result



## driver_code
n = 8
result = findfactorial(n)
print(f"The factorial of {n} is {result}")
```
