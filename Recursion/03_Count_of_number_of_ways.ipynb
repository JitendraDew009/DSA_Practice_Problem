```python
# Overall Time complexity is O(2^n)
# Defination of fucntion : finaboncci series
def helper(n):
    if n<=1:
        return n
    return helper(n-1) + helper(n-2)

# Defination of fucntion   
def countNumberofWay(s):
    return helper(s+1) # why? because our finbonacci series starts from index 2

# Driver_code
n = 5
#Function_calling
result = countNumberofWay(n)
print(f"Number of Ways to reach upstairs for {n} is {result}")
```
