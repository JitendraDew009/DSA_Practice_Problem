```python
# Implemenatio of Selection Procedure
# Time Complexity: 
# Worst Case -> O(n^2)
# Best Case  -> O(n)
# fucntion defination of partiton 
def partition(arr, p, q):
    i = p
    pivot = arr[p]
    for j in range(i+1, q+1):
        if arr[j] <= pivot:
            i+= 1
            arr[i], arr[j] = arr[j], arr[i]
    arr[i], arr[p] = arr[p], arr[i]
    return i+1

# Function Defination
def selection_Procedure(arr, p, q, k):
    if len(arr) == 1:
        return arr[p]
    else:
        # function calling of partition fucntion
        m = partition(arr, p, q)
        if m == k:
            return arr[m-1]
        elif m < k:
            # recursive call -> right side of the pivot element
            return selection_Procedure(arr, m , q, k)
        else:
            # recursive call -> left side of the pivot element
            return selection_Procedure(arr, p, m-2, k)
        

# Driver Code
arr = [20, 39, 40, 47, 50, 65, 70, 75] # Already Sorted Array
#######[1, 2,  3,  4,  5,  6,  7, 8]
k = 4
p = 0
q = len(arr) - 1 
result= selection_Procedure(arr, p, q, k)
print("kth smallest in an array", result)
```
