```python
# Find the Power of Element
# Time Complexity : O(log(n))
## defination of function
def findPowerofElement(a,n):
    # Small problem
    if n == 1:
        return a
    else:
        #big problem -> divide and Conquer Appoach
        mid = n//2
        ## recursive call
        b = findPowerofElement(a,mid)
        result = b * b
        # even
        if n % 2 == 0: 
            return result
        # odd 
        else:
            return result * a
# Driver code
a = 2
n = 10
#function calling
result = findPowerofElement(a,n)
print(f"{a} raised to the Power of {n} is {result}")
```

```python
def findPowerofElement(a, n):
    # Base cases
    if n == 0:
        return 1
    elif n < 0:
        return findPowerofElement(1/a, -n)

    # Exponentiation by squaring
    if n % 2 == 0:  # even exponent
        half = findPowerofElement(a, n // 2)
        return half * half
    else:  # odd exponent
        half = findPowerofElement(a, (n - 1) // 2)
        return a * half * half

# Driver Code
a = 2
n = 10000
result = findPowerofElement(a, n)
print(f"{a} raised to the power {n} is {result}")
```
