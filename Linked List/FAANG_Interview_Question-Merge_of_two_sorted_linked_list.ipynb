```python
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None

class LinkedList:
    def __init__(self):
        self.head = None

    ## insert at the front of the linkedlist
    def insertAtBeginning(self, new_data):
        # creation of the new node
        new_node = Node(new_data)
        
        # insertion at the beginning
        new_node.next = self.head
        self.head = new_node
     
     ## print of the linkedlist
    def printList(self):
        temp =self.head
        while temp:
            print(str(temp.data) + " ", end = " ")
            temp = temp.next

# merge defination to merge the lists
def mergeLists(head1, head2):
    temp = None
    ## llist1 is empty
    if head1 is None:
        return head2
    ## llist2 is empty
    if head2 is None:
        return head1
    
    if head1.data <= head2.data:
        temp = head1
        temp.next = mergeLists(head1.next, head2)
    else:
        temp = head2
        temp.next = mergeLists(head1, head2.next)    
    return temp
## Driver Code
llist1 = LinkedList()

## function calling
print("First Linked List")
llist1.insertAtBeginning(12)
llist1.insertAtBeginning(10)
llist1.insertAtBeginning(9)
llist1.insertAtBeginning(8)
llist1.insertAtBeginning(5)
llist1.printList()

print("\nSecond Linked List")
llist2 = LinkedList()
llist2.insertAtBeginning(14)
llist2.insertAtBeginning(13)
llist2.insertAtBeginning(11)
llist2.insertAtBeginning(7)
llist2.insertAtBeginning(6)
llist2.printList()

print("\nMerged Linked List")

llist3 = LinkedList()
llist3.head = mergeLists(llist1.head, llist2.head)
llist3.printList()
```
